```python
def number(n):
    if n%2 == 0:
        return "Even"
    else:        
        return "Odd"
print(number(7))
```

```python
def number(n):
    if n==1:
        return n
    else:
        return n*number(n-1)
print(number(5))
```

```python
def character(n):
    if n in 'aeiouAEIOU':
        return "Vowel"
    else:
        return "Consonant"
print(character('a'))
```

```python
def character(n):
    if n==n[::-1]:
         return "Palindrome"
    else:
         return "Not a Palindrome"
    
print(character("madam"))
    
```

```python
def number(n):
    if n<=1:
        return "Not prime"
    for i in range(2,int(n*(0.5)+1)):
        if n%i == 0:
            return "Not prime"
    else :
        return "Prime"
print(number(5))
```

```python
def character(n):
    count = 0
    for chr in n:
        if chr in 'aeiouAEIOU':
            count += 1
    return count
print(character("hello world"))
```

```python
def add(a,b):
    a,b=int(a),int(b)
    a,b=b,a
    return a,b
print(add(5,10))
```

```python
def add(a):
   freq={}
   for ch in a:
    freq[ch]=freq.get(ch,0)+1
    return freq
print(add("hello world"))
```

```python
def add(text):
    freq = {}
    for ch in text:
        freq[ch] = freq.get(ch, 0) + 1
    return freq

print(add("hello world"))
```

```python
m=[1,2,3,4,5]
def add(m):
    n=[]
    v=[]
    for i in m:
        if i%2==0:
            n.append(i)
        else:
            v.append(i)
    return n,v
print(add(m))
```
