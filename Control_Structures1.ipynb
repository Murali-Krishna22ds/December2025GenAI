```python
x=[11,22,3,2244,45,61,72,822,91,102]
even_sum=0
odd_sum=0
var=0
while var<10:
  if(x[var]%2==0):
    even_sum=even_sum+x[var]
  else:
    odd_sum=odd_sum+x[var]
  var=var+1
print(even_sum,odd_sum)
```
