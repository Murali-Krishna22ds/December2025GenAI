```python
x=97
is_prime = True

if x <= 1:
    is_prime = False
else:
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{x} is a Prime Number")
else:
    print(f"{x} is NOT a Prime Number")
```
